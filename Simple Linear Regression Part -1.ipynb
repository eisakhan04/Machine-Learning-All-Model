Ordinary Least Squares regression (OLS) is a common technique for estimating coefficients of linear regression equations which describe the relationship between one or more independent quantitative variables and a dependent variable (simple or multiple linear regression).



# Example 01

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns 
```

```python
df= pd.DataFrame({'cgpa':[2,3,4,5,6],'package':[3.2,4.1,5.1,4.1 ,3.2]})
df
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
sns.scatterplot(x="cgpa" , y="package" ,data= df)
plt.show()
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1] 
```

```python
y.ndim
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr
```

```python
lr.fit(X_train,y_train)
```

```python
X_train.iloc[0]
```

```python
lr.predict(X_train.iloc[0].values.reshape(1,1))
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
# 2	3.2
# 1	3	4.1
# 2	4	5.1
# 3	5	4.1
# 4	6	3.2

lr.predict([[5]])
```

```python
predict = lr.predict(X)
```

```python
sns.scatterplot(x="cgpa" , y="package" ,data= df)
plt.plot(df["cgpa"] , predict , c="k")
plt.legend(["org Data" ,"predict_Line"] )
plt.show()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
m = lr.coef_
m
```

```python
b = lr.intercept_
b
```

```python
lr.score(X_test , y_test)
```
