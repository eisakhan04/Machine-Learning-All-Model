# Types of Naive bayes Models:
* there are three types of Naive bayes Models is following 
* Gaussian      : features follow a Gaussian(normal) distribution input data
* Multinomial   : features follow a multinomial distribution and such as work on textual Data 
* Bernoulli     : features are follow that binary (Boolean) variables and classification problem 

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
datas =  pd.read_csv("ML Algorithms/Social_Network_Ads (2).csv")
```

```python
datas.head()
```

```python
sns.kdeplot(data=datas["Age"])
plt.show()
```

```python
sns.kdeplot(data=datas["EstimatedSalary"])
plt.show()
```

```python
datas.isnull().sum()
```

```python
sns.scatterplot(x= "Age" , y="EstimatedSalary" , data= datas , hue="Purchased")
plt.show()
```

```python
x = datas.iloc[ : , : -1]
y = datas["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x  , y , test_size=0.2 , random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BrenoulliNB
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
```

# Gaussian naive_bayes algorithm 

```python
gnb = GaussianNB()
gnb.fit( x_train, y_train)
```

```python
gnb.score(x_test , y_test)*100
```

```python
gnb.score(x_train, y_train)*100
```

```python
gnb.predict([[27 , 57000]])
```

```python
gnb.predict(x_test)
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=gnb)
plt.show()
```

# Multinomial naive_bayes algorithm 

```python
mnb = MultinomialNB()
mnb.fit(x_train , y_train)
```

```python
mnb.score(x_test , y_test)*100
```

```python
mnb.score(x_train , y_train)*100
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=mnb)
plt.show()
```

# Bernoulli naive_bayes algorithm 

```python
bnb =BernoulliNB()
bnb.fit(x_train , y_train)
```

```python
bnb.score(x_test , y_test)
```

```python
bnb.score(x_train , y_train)
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=bnb)
plt.show()
```
