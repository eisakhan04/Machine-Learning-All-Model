# type of logistic Trgression

* Binomial: there can be only two possible types of the dependent variables such as : true or false,  0 or 1 , fail or pass.
* Multinomial: there can be 3 or more possible unorder type dependent variables , example : "cat" , "dogs" , "sheep".
* Ordinal: there can be 3 oor more possible order type dependent variables , example : "Low" , "Medium" , "High".
